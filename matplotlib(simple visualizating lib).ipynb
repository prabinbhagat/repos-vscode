# MatPlotLib 

Matplot lib is a plotting library for Python programming language and its numerical mathematical extension NumPY.   
It provides an object oriented API for embedding plots into applicationns using general-purpose GUI toolkits 
like Tkinter,wxPython,Qt or GTK+

Some of major Pros of MatplotLib are
1.  Generally easy to get started for simple plots
2.  Support for custorm label and text.
3.  Great conrol of every element in a figure.
4.  High Quality output in many formats.
5.  Customizable in general.

```python
# Sub Plots withing one diagram i can create multiple subplot withing the canvas.

plt.subplot(2,2,1) # row,columns, index = position  ''row and coloumn provides the size of the canvas
plt.plot(x,y,1)

plt.subplot(2,2,2)     # last argument is the positon where i want to plot
plt.plot(x,y,'r--',2)

plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
# Compute the x and y coordinate for points on a sine wave
x = np.arange(0,4*np.pi,0.1)    # here third parameter is step 
y = np.sin(x)
plt.plot(x,y)
plt.title("Sine Wave")
```

```python
# Subplot
# Compute the x and y coordinates for points on sine and cosine curve

x = np.arange(0,5*np.pi,0.1)
ySin = np.sin(x)
yCos = np.cos(x)

# Now set up a subplot gird with height 2 and width 1
# and set the first such plot as active
plt.subplot(2,1,1)          # it will create 2 rows with one column

# Make the first plot
plt.plot(x,ySin)
plt.title('Sine Wave')

# Set the second subplot as active and make the second plot
plt.subplot(2,1,2)
plt.plot(x,yCos,'r--')
plt.title("Cos Wave")
```

```python
## Histograms
#   it helps us to find out wrt numbers what is the density , what is the count that we can see in y axix.
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])

plt.hist(a)  # by default it is 10 bins
plt.title("Histogram")
```

```python
## Pie Chart

# Data to plot
labels = 'Python','C++','C#','VB.net'    # Just the labels
sizes = [215,130,245,210]               # it will assign as %age internally
colors = ['gold','yellowgreen','lightcoral','lightskyblue']     #just the colors
explode = [0.2,0.0,0.0,0.0]              # how much far the slices are to move far away

#plot
plt.pie(sizes,explode,labels,colors,
autopct = '%1.1f%%',shadow =True )      # autopercent shows in what format eg. 1.1 floating format and shadow will provide some shadwo
plt.axis('equal')
```
